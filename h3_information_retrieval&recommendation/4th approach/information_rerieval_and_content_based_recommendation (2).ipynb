```python
import pandas as pd
import spacy
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: Load and preprocess the dataset
data = pd.read_csv("model3.csv")

# Step 2: Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Step 3: Create a TF-IDF vectorizer for information retrieval
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(data['text'])

# Step 4: Compute cosine similarities for information retrieval
cosine_similarities = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Step 5: Information Retrieval
# You can retrieve documents based on user queries
def retrieve_documents(query, n=10):
    query_vector = tfidf_vectorizer.transform([query])
    cosine_scores = cosine_similarity(query_vector, tfidf_matrix)
    top_indices = cosine_scores.argsort()[0][::-1][:n]
    return data.iloc[top_indices]

# Step 6: Content-Based Recommendation
def recommend_similar_documents(query, n=5):
    query_vector = nlp(query).vector
    similarity_scores = []

    for doc in data['text']:
        doc_vector = nlp(doc).vector
        similarity = cosine_similarity([query_vector], [doc_vector])
        similarity_scores.append(similarity[0][0])

    top_indices = sorted(range(len(similarity_scores)), key=lambda i: similarity_scores[i], reverse=True)[:n]
    recommendations = data.iloc[top_indices]
    return recommendations

# Step 7: User Interface
def main():
    while True:
        print("Welcome to the Recommendation System")
        print("1. Retrieve Documents")
        print("2. Content-Based Recommendations")
        print("3. Exit")

        choice = input("Enter your choice (1/2/3): ")

        if choice == '1':
            query = input("Enter your query: ")
            n = int(input("Enter the number of documents to retrieve: "))
            retrieved_documents = retrieve_documents(query, n)
            print("Retrieved Documents:")
            for i, row in retrieved_documents.iterrows():
                print(f"{i + 1}. {row['text']}")

        elif choice == '2':
            query = input("Enter a document text for which you want recommendations: ")
            n = int(input("Enter the number of recommendations to retrieve: "))
            recommendations = recommend_similar_documents(query, n)
            print("Content-Based Recommendations:")
            for i, row in recommendations.iterrows():
                print(f"{i + 1}. {row['text']}")

        elif choice == '3':
            print("Goodbye!")
            break

        else:
            print("Invalid choice. Please enter a valid option (1/2/3).")

if __name__ == "__main__":
    main()
```
